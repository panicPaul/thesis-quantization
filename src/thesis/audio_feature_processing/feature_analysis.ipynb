```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from thesis.data_management import SequenceManager
from tqdm import tqdm

sm = SequenceManager(3, cleaned_audio=False)

video = []
for i in tqdm(range(len(sm))):
    vector = sm.flame_params[i]

    # Create a new figure for each frame
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_ylim(-10, 10)
    ax.plot(vector)

    # Convert plot to image
    fig.canvas.draw()
    frame = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
    frame = frame.reshape(fig.canvas.get_width_height()[::-1] + (4,))
    frame = frame[:, :, :-1]

    # Convert RGB to BGR (OpenCV uses BGR)
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    video.append(frame)

    # Close the figure to free up memory
    plt.close(fig)

# Define the output video path
output_path = "tmp/audio_features.mp4"

# Get the frame size
height, width, _ = video[0].shape

# Create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, 15, (width, height))

# Write frames to video
for frame in video:
    out.write(frame)

# Release the VideoWriter
out.release()

print(f"Video saved to {output_path}")
```

```python
# together
import matplotlib.pyplot as plt
import cv2
import numpy as np
from thesis.data_management import SequenceManager
from tqdm import tqdm

sm_raw = SequenceManager(3, cleaned_audio=False)
sm_clean = SequenceManager(3, cleaned_audio=True)

video = []
for i in tqdm(range(len(sm_raw))):
    vector_raw = sm_raw.audio_features[i]
    vector_clean = sm_clean.audio_features[i]

    # Create a new figure for each frame
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_ylim(-10, 10)
    ax.plot(vector_raw)
    ax.plot(vector_clean)

    # Convert plot to image
    fig.canvas.draw()
    frame = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
    frame = frame.reshape(fig.canvas.get_width_height()[::-1] + (4,))
    frame = frame[:, :, :-1]

    # Convert RGB to BGR (OpenCV uses BGR)
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    video.append(frame)

    # Close the figure to free up memory
    plt.close(fig)

# Define the output video path
output_path = "tmp/audio_features.mp4"

# Get the frame size
height, width, _ = video[0].shape

# Create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, 15, (width, height))

# Write frames to video
for frame in video:
    out.write(frame)

# Release the VideoWriter
out.release()

print(f"Video saved to {output_path}")
```
